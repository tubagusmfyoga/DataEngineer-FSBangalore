## **Menggabungkan Data**

Proses atau kegiatan menyatukan data dari berbagai sumber atau tabel berdasarkan hubungan tertentu agar menjadi satu informasi utuh, dengan tujuan memudahkan analisis, pengambilan keputusan, dan pemahaman data secara menyeluruh.

Dua buah data atau lebih bisa digabungkan **asalkan memiliki kolom kunci (key) yang sama atau saling berhubungan**.

Cara menggabungkan data di pandas bisa dilakukan dengan mengeksekusi perintah berikut

```
DataFrame1.merge(DataFrame2, left_on = [key_df1], right_on = [key_df2], how)
```

<br>Dengan ketentuan :     
* **DataFrame1** adalah dataframe disebelah kiri
* **DataFrame2** adalah dataframe disebelah kanan
* **key_df1** adalah kunci pada DataFrame1 yang akan digabungkan
* **key_df2** adalah kunci pada DataFrame2 yang akan digabungkan  
* **how** adalah cara penggabungan yakni bisa ‘left’, ‘right’, ‘outer’, ‘inner’, ‘cross‘ secara default adalah ‘inner’


<br>docs : *https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.merge.html*

### 👨‍💻 Task 4.1

ERD (Entity Relationalship Diagram) pada DataFrame transaksi pizza disajikan oleh gambar berikut

<img src="https://raw.githubusercontent.com/bachtiyarma/Material/refs/heads/main/Image/Materi-SQL/ERD%20-%20Pizza.png" width="60%">

Buat DataFrame baru bernama `pizza_transaction` yang merupakan gabungan dari DataFrame yang telah dibuat sebelumnya yakni DataFrame `orders` dan `order_details`


````python
# Lakukan proses penggabungan data
pizza_transaction = orders.merge(order_details, on = 'order_id', how = 'left')

# Tampilkan sampel datanya
display(pizza_transaction)
````

### 👨‍💻 Task 4.2

Kemudian hubungkan DataFrame `pizza_transaction` dengan DataFrame `customers`, `pizzas` dan `pizza_types` untuk mendapatkan informasi detail mengenai produk pizza yang terjual


````python
# Lakuka proses penggabungan data
pizza_transaction =pizza_transaction.merge(customers, on='customer_id', how='left')\
                                    .merge(pizzas, on='pizza_id', how='left')\
                                    .merge(pizza_types, on='pizza_type_id', how='left')

# Tampilkan sampel datanya
display(pizza_transaction)
````

### 👨‍💻 Task 5.2

Kolom `price`	dan `production_cost` pada DataFrame `pizza_transaction`, masih mengandung simbol Currency. Hapus simbol tersebut dan konversi tipe datanya menjadi float

<br>docs : *https://pandas.pydata.org/docs/reference/api/pandas.Series.str.replace.html*

````python
pizza_transaction.info()
````

````python
# Pastikan kolom price diubah ke string dulu, hapus IDR, lalu konversi ke Float64Dtype (nullable float)
pizza_transaction['price'] = pizza_transaction['price'].astype(str).str.replace('IDR', '')
pizza_transaction['price'] = pizza_transaction['price'].astype(pd.Float64Dtype())

# Periksa hasilnya
display(pizza_transaction)
pizza_transaction.info()
````

````python
# Hapus simbol currency pada kolom production_cost
pizza_transaction['production_cost'] = pizza_transaction['production_cost'].str.replace('IDR', '')

# Konversi kolom production_cost menjadi float
pizza_transaction['production_cost'] = pizza_transaction['production_cost'].astype(pd.Float64Dtype())

# Periksa hasilnya
display(pizza_transaction)
````

##**Aggregation**

Agregasi data merupakan serangkaian proses yang dilakukan untuk melakukan pengumpulan data dan dinyatakan dalam bentuk ringkasan sebagai contoh menghitung total yang dikelompokkan berdasarkan kategori tertentu, dsb.

Untuk melakukan agregasi gunakan sintaks berikut

<p align="center"><code><b>DataFrame.groupby(by, as_index).agg(kolom_hasil_aggregasi = (kolom_dihitung, jenis_agregasi))</b></code></p>

Jenis agregasi :
* sum = Menghitung total data
* mean = Menghitung rataan data
* max = Menghitung nilai maksimum data
* min = Menghitung nilai minimum data
* nunique = Menghitung jumlah data unik data
* count = Menghitung banyak data

<br>docs : *https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.groupby.html*

### 👨‍💻 Task 7

Pada DataFrame `pizza_transaction`, lakukan perhitungan :      
* Total kuantitas
* Total amount (price)
* Total profit
* Total komplain

untuk setiap harinya pada kategori dan nama pizza terjual. Simpan hasil perhitungan kedalam dataframe `result`

````python
# Lakukan perhitungan agregasi
result = pizza_transaction.groupby(by = ['order_date', 'pizza_name', 'pizza_category'], as_index = False).agg(
    total_kuantitas = ('quantity', 'sum'),
    total_amount = ('total_price', 'sum'),
    total_profit = ('total_profit', 'sum'),
    total_complain = ('is_complain', 'sum')
)

# Tampilkan hasilnya
display(result)
````
