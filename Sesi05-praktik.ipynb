```python
import psycopg2

# connect ke postgres default
conn = psycopg2.connect("dbname=postgres user=postgres password=postgres123 host=localhost")
conn.autocommit = True
```

```python
# import data csv
import pandas as pd

df = pd.read_csv("sesi05-sales_with_json.csv")
print(df.head())
```

```python
# import data csv
import pandas as pd

df = pd.read_excel("sales_with_json.xlsx")
print(df.head())
```

```python
# Buat tabel sales
with conn.cursor() as cur:
    cur.execute("""
        CREATE TABLE IF NOT EXISTS sales_with_json (
            order_id INT PRIMARY KEY,
            customer_id INT,
            city TEXT,
            quantity INT,
            price NUMERIC,
            order_details JSONB
        );
    """)
conn.commit()
```

```python
# Insert data CSV ke tabel sales
with conn.cursor() as cur:
    for _, row in df.iterrows():
        cur.execute("""
            INSERT INTO sales_with_json (order_id, customer_id, city, quantity, price, order_details)
            VALUES (%s, %s, %s, %s, %s, %s)
        """, (
            row['order_id'],
            row['customer_id'],
            row['city'],
            row['quantity'],
            row['price'],
            row['order_details']   # sudah format JSON-string di CSV
        ))
conn.commit()
```
