# Gunakan fitur Search dari Library Googleplay Scraper untuk search data yang mau scraping dan tampilkan Hasilnya

```python
#Melakukan Search Data dari "Bank Syariah"
from google_play_scraper import search

results=search(
    "bank syariah",
    lang="id",
    country="id",
    n_hits=50
)

#Menampilkan Hasil
display(results)
```

# Cek Tipe Data dan Panjang Data

```python
#Mengecek tipe data Hasil
print(type(results))

#Mengecek panjang data Hasil
print(len(results))
```

# Menampung Hasil scraping ke variable baru bernama "app_id"

```python

app_id = [elemen.get('appId') for elemen in results]
print(app_id)
```

# Menampung Hasil scraping ke Variable baru bernama "app_title"

```python
app_title = [elemen.get('title') for elemen in results]
print(app_title)
```

# Gunakan Fitur review_all dari Library Googleplay Scraper untuk ambil data Semua Review dan Libray Pandas untuk olah data

```python
#Melakukan pencarian Semua Review yang ada di semua variable app_id
from google_play_scraper import reviews_all
import pandas as pd

#buat Data Frame  Kosong
data_itegrasi = pd.DataFrame()

#Proses Ekstrasi Review
for id, title in zip(app_id, app_title):
    reviews = reviews_all (
        id,
        sleep_milliseconds=100,
        lang='id',
        country='id'
    )

  #Ubah ke Data frame
    data_reviews = pd.DataFrame(reviews)

  #Buat Kolom Baru Sebagai Penanda
    data_reviews['app_id'] = id
    data_reviews['app_title'] = title

  ##Proses Append Data
    data_itegrasi = pd.concat([data_itegrasi, data_reviews])
```

# Buat DataSet Baru bernama review_gplay_bank_syariah

```python
from google.cloud.bigquery import dataset
#Inisialisasi project_id
project_id = 'fullstack-bangalore-494812'
client = bigquery.Client( project = project_id)

#Buat DataSet baru
dataset_id = 'review_gplay_bank_syariah'
dataset_ref = bigquery.Dataset(f'{project_id}.{dataset_id}')
dataset = client.create_dataset(dataset_ref, exists_ok=True)
```

```python
to_gbq(
  data_itegrasi,
  destination_table = f'{dataset_id}.review_all_bank_syariah',
  project_id = project_id,
  if_exists = 'append')
```
